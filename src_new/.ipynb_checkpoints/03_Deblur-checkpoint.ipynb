```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import convolve2d as conv2
from skimage import color, data, restoration
```

```python
img = cv2.imread("/home/damvantai/Pictures/smplayer_screenshots/cap_1_03_H_082018120000_00:06:29_02.jpg")
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
gray.shape
```

```python
psf = np.ones((5, 5)) / 25
gray = conv2(gray, psf, 'same')
img_noise = gray.copy()
img_noise += (np.random.poisson(lam=25, size=gray.shape) - 10) / 255.
deconvolved_RL = restoration.richardson_lucy(img_noise, psf, iterations=30)
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(8, 5))
plt.gray()

for a in (ax[0], ax[1], ax[2]):
       a.axis('off')

ax[0].imshow(gray)
ax[0].set_title('Original Data')

ax[1].imshow(img_noise)
ax[1].set_title('Noisy data')

ax[2].imshow(deconvolved_RL, vmin=img_noise.min(), vmax=img_noise.max())
ax[2].set_title('Restoration using\nRichardson-Lucy')


fig.subplots_adjust(wspace=0.02, hspace=0.2,
                    top=0.9, bottom=0.05, left=0, right=1)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.signal import convolve2d as conv2

from skimage import color, data, restoration

# astro = color.rgb2gray(data.astronaut())
img = cv2.imread("/home/damvantai/Pictures/smplayer_screenshots/cap_1_03_H_082018120000_00:06:29_02.jpg")
astro = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
psf = np.ones((5, 5)) / 25
astro = conv2(astro, psf, 'same')
# Add Noise to Image
astro_noisy = astro.copy()
astro_noisy += (np.random.poisson(lam=25, size=astro.shape) - 10) / 255.

# Restore Image using Richardson-Lucy algorithm
deconvolved_RL = restoration.richardson_lucy(astro_noisy, psf, iterations=30)

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(8, 5))
plt.gray()

for a in (ax[0], ax[1], ax[2]):
       a.axis('off')

ax[0].imshow(astro)
ax[0].set_title('Original Data')

ax[1].imshow(astro_noisy)
ax[1].set_title('Noisy data')

ax[2].imshow(deconvolved_RL, vmin=astro_noisy.min(), vmax=astro_noisy.max())
ax[2].set_title('Restoration using\nRichardson-Lucy')


fig.subplots_adjust(wspace=0.02, hspace=0.2,
                    top=0.9, bottom=0.05, left=0, right=1)
plt.show()
```
